````python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
````

````python
def add_user_message(messages, text):
    user_message={"role":"user","content":text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message={"role":"assistant","content":text}
    messages.append(assistant_message)

def chat(messages, system=None):
    params={
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    
    if system:
        params["system"]= system

    message=client.messages.create(**params)
    return message.content[0].text
````

````python
# 실습문제3. stop sequence와 message prefilling 활용하여 구조화된 답안 뽑아내기
# 주의!) 강의는 sonnet 4-0버전으로 촬영되었으나,학습중 사용한 최신 버전인 sonnet 4-6에서는 모종의 사유로 message prefilling 기능이 더 이상 지원되지 않음이 확인됨
# 따라서 message prefilling 대신 시스템 프롬프트를 통해서 예제를 풀이함
messages = []
prompt = """
AWS CLI 명령어의 세가지 다른 샘플을 뽑아내어라. 각각의 샘플 길이는 매우 짧아야 한다."""

add_user_message(messages, prompt)
text = chat(messages)
text.strip()
text
````

````python
from IPython.display import Markdown

Markdown(text)
````
